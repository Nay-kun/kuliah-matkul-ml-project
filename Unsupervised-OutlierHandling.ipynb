```python
import pandas as pd
import numpy as np

# read data
data = pd.read_csv('data-project-ml-binned.csv')
data.head()
```

```python
data.info()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame and it's already loaded

# 1. Scale the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 2. Apply DBSCAN
# Adjust these parameters based on your data
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(data_scaled)

# 3. Identify outliers
data['dbscan_label'] = dbscan.labels_

# Points with label -1 are considered outliers
outliers = data[data['dbscan_label'] == -1]

print("Outlier Rows:")
print(outliers)
```
